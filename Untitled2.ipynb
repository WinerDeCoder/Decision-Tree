```python
import torch

import numpy as np
```

```python
dataset = torch.randn(30,5)
print(dataset)
```

```python
label = torch.zeros(30,1)
index_1 = [1,3,5,7,9,11,13,14,16,18,20,22,24,26,29,25]
for index in index_1:
  label[index] = 1

print(label)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model = DecisionTreeClassifier(criterion="gini",max_depth = 3, random_state = 1)

tree_model.fit(dataset,label)
```
